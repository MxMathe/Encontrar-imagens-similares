```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import requests
from tqdm import tqdm
import zipfile
import webbrowser
```

Extração arquivos .zip

```python
zip_path = "/content/archive (1).zip"
extract_path = "/content/dataset"

os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"Dataset extraído para: {extract_path}")
```

```python
# Caminho do arquivo CSV do dataset
csv_path = "/content/dataset/shoes_dim.csv"

# Diretório para salvar as imagens
output_dir = "imagens_adidas"
os.makedirs(output_dir, exist_ok=True)

# Carregar o dataset
df = pd.read_csv(csv_path)

image_column = "image_url"

# Baixar as imagens
for idx, url in tqdm(enumerate(df[image_column]), total=len(df)):
    try:
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            image_path = os.path.join(output_dir, f"imagem_{idx}.jpg")
            with open(image_path, "wb") as file:
                for chunk in response.iter_content(1024):
                    file.write(chunk)
    except Exception as e:
        print(f"Erro ao baixar {url}: {e}")

print("Download concluído!")
```

DATASET

```python
dataset_path = "/content/imagens_adidas"
```

```python
model = ResNet50(weights="imagenet", include_top=False, pooling="avg")
```

```python
features = []
image_paths = []

# Função para carregar e processar imagens
def extract_features(img_path):
    img = image.load_img(img_path, target_size=(224, 224))  # Redimensiona
    img_array = image.img_to_array(img)                     # Converte para array
    img_array = np.expand_dims(img_array, axis=0)           # Adiciona batch dimension
    img_array = preprocess_input(img_array)                 # Normaliza pixels
    feature_vector = model.predict(img_array)               # Extrai features
    return feature_vector.flatten()
```

```python
for img_name in tqdm(os.listdir(dataset_path)):
    img_path = os.path.join(dataset_path, img_name)
    features.append(extract_features(img_path))
    image_paths.append(img_path)
```

```python
# Converte lista de features para numpy array
features = np.array(features)

# Cria o modelo KNN para busca de similaridade
knn = NearestNeighbors(n_neighbors=5, metric="euclidean")
knn.fit(features)
```

```python
# Função para encontrar imagens similares
def find_similar_images(query_image_path, k=5):
    query_features = extract_features(query_image_path).reshape(1, -1)
    distances, indices = knn.kneighbors(query_features, n_neighbors=k)

    # Mostra a imagem de entrada
    plt.figure(figsize=(15, 5))
    plt.subplot(1, k+1, 1)
    plt.imshow(image.load_img(query_image_path))
    plt.title("Imagem de Entrada")
    plt.axis("off")

    # Mostra as imagens similares
    for i, index in enumerate(indices[0]):
        plt.subplot(1, k+1, i+2)
        plt.imshow(image.load_img(image_paths[index]))
        plt.title(f"Similar {i+1}")
        plt.axis("off")

    plt.show()
```

```python
query_image_path = "/content/1-3963.jpg"
find_similar_images(query_image_path)
```
